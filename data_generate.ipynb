```python
classes = ['person', 'bus', 'truck', 'car', 'bicycle', 'motorbike', 'cat', 'dog', 'horse', 'sheep', 'cow']
class_shape = [[175, 55, 30], [300, 250, 1200], [300, 250, 1200], [160, 180, 400], [110, 50, 180],
                [110, 50, 180], [40, 20, 50], [50, 30, 60], [180, 60, 200], [130, 60, 150], [170, 70, 200]]
class_size = dict(zip(classes,class_shape))
```

```python
class_size
```

```python
image_annotations={}
for image,data in box_data.items():
    image_annotation={}
    image_annotation["name"]= data["filename"]
    image_annotation["class"]='bicycle'
    image_annotation["size_h"]=class_size[image_annotation["class"]][0]
    image_annotation["size_w"]=class_size[image_annotation["class"]][1]
    image_annotation["size_d"]=class_size[image_annotation["class"]][2]
    image_annotation["distance"]=dist_data[image_annotation["name"]]
    box_region =data["regions"]
    if box_region.keys():
        box_attributes = box_region['0']
        box_shape = box_attributes['shape_attributes']
        image_annotation["height"]=box_shape["height"]
        image_annotation["width"]=box_shape["width"]
        image_annotation["x"]=box_shape["x"]
        image_annotation["y"]=box_shape["y"]
        image_annotations[image_annotation["name"]]=image_annotation    
```

```python
image_annotations
```

```python
with open("test_data_2/2018-05-23-16-38-11_03_final_annotation.json", "w")as file:
    json.dump(image_annotations,file)
```

```python
final_annotation_files={}

files = os.listdir("../data_distance")
for file in files:
    if file.endswith("annotation.json"):
        with open(file)as f:
            data=json.load(f)
            print(len(data))
            final_annotation_files=dict(final_annotation_files.items()+data.items())
 
```

```python
def load_data(data_dict,img_width,img_height):
    data_array=[]
    output_array=[]
    for data in final_annotation_files.values():
        width = float(data['width'])/img_width
        height = float(data['height'])/img_height
        x1=float(data['x'])/img_width
        y1=float(data['y'])/img_height
        x2=float(data['x']+data['width'])/img_width
        y2=float(data['y'])/img_height
        x3=float(data['x'])/img_width
        y3=float(data['y']+data['height'])/img_height
        x4=float(data['x']+data['width'])/img_width
        y4=float(data['y']+data['height'])/img_height
        class_h = float(data["size_h"])/100
        class_w = float(data["size_w"])/100
        class_d = float(data["size_d"])/100
        current_object=[width,height,x1,x2,x3,x4,y1,y2,y3,y4,class_h,class_w,class_d]
        data_array.append(current_object)
        output_array.append(data["distance"])
    return np.array(data_array),np.array(output_array)
```
